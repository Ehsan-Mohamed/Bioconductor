```r
# Extract probe IDs (Affymetrix IDs)
probes = featureNames(ALL)
```

```r
# Connect to Ensembl 75 (hg19 build)
ensembl75 = useMart(
  host = "feb2014.archive.ensembl.org",
  biomart = "ENSEMBL_MART_ENSEMBL"
)

# Choose the correct dataset for humans
ensembl75 = useDataset("hsapiens_gene_ensembl", mart = ensembl75)

# Annotate: Map Affymetrix probe IDs to Ensembl Gene IDs
annot = getBM(
  attributes = c("affy_hg_u95av2", "ensembl_gene_id"),
  filters = "affy_hg_u95av2",
  values = probes,
  mart = ensembl75
)

```

```r
# Remove rows with missing annotations
annot <- na.omit(annot)

# Count how many probes map to more than one Ensembl gene
multi_map_count <- sum(duplicated(annot))

# Or more clearly:
table_counts <- table(annot$affy_hg_u95av2)
multi_gene_probes <- sum(table_counts > 1)

# Output the result
cat("Number of probesets annotated with more than one Ensembl gene ID:", multi_gene_probes, "\n")
```

```r
# Choose human gene dataset
ensembl75 <- useDataset("hsapiens_gene_ensembl", mart = ensembl75)

# Retrieve annotations including chromosome
annot <- getBM(
  attributes = c("affy_hg_u95av2", "ensembl_gene_id", "chromosome_name"),
  filters = "affy_hg_u95av2",
  values = probes,
  mart = ensembl75
)

# Filter only for autosomal chromosomes 1 through 22
autosomes <- as.character(1:22)
annot_autosomes <- annot[annot$chromosome_name %in% autosomes, ]

# Count unique probesets with at least one autosomal gene annotation
unique_probes_autosomes <- unique(annot_autosomes$affy_hg_u95av2)
num_probes_autosomes <- length(unique_probes_autosomes)

# Output the result
cat("Number of probesets annotated with one or more genes on autosomes (chr 1–22):", num_probes_autosomes, "\n")
```

```r
# Load the MsetEx dataset
data(MsetEx)

# Check sample names to ensure the target sample is present
sampleNames(MsetEx)

# Extract the methylated signal matrix
meth_matrix <- getMeth(MsetEx)

# Confirm target sample exists
"5723646052_R04C01" %in% colnames(meth_matrix)

# Calculate the mean methylation value across all features for the target sample
mean_methylation <- mean(meth_matrix[, "5723646052_R04C01"], na.rm = TRUE)

# Output the result
cat("Mean methylation value for sample 5723646052_R04C01:", mean_methylation, "\n")
```

```r
# Define an S4 class called "Person"
setClass("Person",
         slots = list(name = "character", age = "numeric"))

# Create an instance of the Person class
john <- new("Person", name = "John Doe", age = 30)

# Access the slots
john@name  # Output: "John Doe"
john@age   # Output: 30
```

```r
# Define an S4 class with a validity check
setClass("Employee",
         slots = list(name = "character", salary = "numeric"),
         validity = function(object) {
           if (object@salary < 0) {
             return("Salary must be non-negative")
           }
           TRUE
         })

# Create a valid Employee object
emp1 <- new("Employee", name = "Alice", salary = 50000)

# Validate the object
validObject(emp1)  # Output: TRUE

# Create an invalid Employee object
#emp2 <- new("Employee", name = "Bob", salary = -1000)

# Validate the invalid object
#validObject(emp2)  # Output: Error: Salary must be non-negative
```
